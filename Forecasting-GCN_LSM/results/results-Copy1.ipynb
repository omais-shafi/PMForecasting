```python
def data_preparation(datafile):
    #Put the file location
    df = pd.read_csv(datafile)
    #type casting
    df.pm1_0 = df.pm1_0.astype(float)
    df.pm2_5 = df.pm2_5.astype(float)
    df.pm10 = df.pm10.astype(float)
    # rounding @15min
    df.dateTime = pd.to_datetime(df.dateTime)
    df.dateTime = df.dateTime.dt.round('15min')
    df.dateTime = df.dateTime.dt.hour*60 + df.dateTime.dt.minute
    # 3pm-12am
    df = df[(df.dateTime>=900)] 
    df = df[(df.dateTime<=1440)] 
    df = df[['pm2_5', 'dateTime']]
    df = df.groupby(['dateTime']).mean()
    df = df.reset_index()
    return df
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import time
import pandas as pd

df = pd.read_csv('out_grid_sizes.csv')
df = df[['GCN sizes', 'LSTM sizes', 'Last Test loss']]
means = df.groupby(['GCN sizes', 'LSTM sizes']).mean()
stds = df.groupby(['GCN sizes', 'LSTM sizes']).std()
sizes = [
            (    '[4, 4]',    '[4]'),
            (    '[4, 4]',    '[8]'),
            (    '[4, 4]',   '[16]'),
            (    '[8, 8]',    '[4]'),
            (    '[4, 4]',   '[32]'),
            (  '[16, 16]',    '[4]'),
            (  '[32, 32]',    '[4]'),
        ]
means = means.loc[sizes]['Last Test loss'].values.tolist()
stds = stds.loc[sizes]['Last Test loss'].values.tolist()
sizes = [
            '([4, 4], [4])',
            '([4, 4], [8])',
            '([4, 4], [16])',
            '([8, 8], [4])',
            '([4, 4], [32])',
            '([16, 16], [4])',
            '([32, 32], [4])',
        ]
train_times = [623.55, 663.05, 749.31, 900.75, 912.92, 1478.27, 4853.53]

fig,ax = plt.subplots(figsize=(9, 5))
fig = plt.figure()
# l1 = ax.plot(sizes, means, color="red", marker="^",linewidth="2", label = 'RMSE')
l1 = ax.errorbar(sizes, means, stds, fmt='ok', capsize=5, ecolor="red", label = 'RMSE')

ax.set_xlabel("Architecture: (GCN layer, LSTM layer)", fontsize=15)
ax.set_ylabel("RMSE", fontsize=15)


ax2=ax.twinx()
l2 = ax2.plot(sizes, train_times,color="blue",marker="o", linewidth="2",linestyle="dashed", label='Runtime')
ax2.set_ylabel("Secure Runtime for 1 epoch (secs)", fontsize=15)
ax2.set_yscale("log")
# Adding legend
# ax.legend(handles=l1+l2)
ax.legend(loc = 2)
ax2.legend(loc = 0, bbox_to_anchor=(0.18,0.9))


plt.setp(ax.get_xticklabels())
plt.setp(ax.get_yticklabels(), fontsize=15)
plt.setp(ax2.get_yticklabels(), fontsize=15)
# plt.savefig("arch_fig.pdf")
plt.show()
```
